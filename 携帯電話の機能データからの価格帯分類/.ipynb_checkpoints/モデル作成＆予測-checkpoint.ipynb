```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
```

```python
df_train = pd.read_csv("./train.csv")
df_test = pd.read_csv("./test.csv")
df_result = pd.read_csv("./sample_submission.csv",header=None)
display(df_train.head())
display(df_test.head())
display(df_result.head())

print(df_train.shape)
print(df_test.shape)
print(df_result.shape)
```

```python
target = ["price_range"]
feature =["blue","fc","int_memory","m_dep","mobile_wt","n_cores","pc","ram","dual_sim","touch_screen","wifi"]
y_train = df_train[target]
X_train = df_train[feature]
model = LogisticRegression()
model.fit(X_train,y_train)
pred = model.predict(X_train)
df_train["pred_price_range"]=pred
display(df_train[["price_range","pred_price_range"]])
display(y_train)
display(pred)
```

```python
f_one = f1_score(pred,y_train,average=None)
print(f_one)
```

```python
X_test = df_test[feature]
pred = model.predict(X_test)
df_result[1] = pred
display(df_result)
```

```python
df_result.to_csv("./result.csv",header=None,index=False)
```
