```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression as LR
from sklearn.metrics import accuracy_score
```

```python
df_train = pd.read_csv("train.csv")
display(df_train.head())
display(df_train.dtypes)
display(df_train.shape)
#display(pd.get_dummies(df_train["Gender"]))
```

```python
df_test = pd.read_csv("test.csv")
display(df_test.head())
display(df_test.dtypes)
display(df_test.shape)
#display(pd.get_dummies(df_test"Gender"]))
df_result = pd.read_csv("sample_submit.csv",header=None)
display(df_result.head())
```

```python
target = ["Attrition"]
features = ["Age","DistanceFromHome","EnvironmentSatisfaction","JobSatisfaction","MonthlyIncome","NumCompaniesWorked","StandardHours"]
y_train = df_train[target]
X_train = df_train[features]
display(y_train.head())
display(X_train.head())
```

```python
clf = LR()
clf.fit(X_train,y_train)
y_pred = clf.predict(X_train)
display(y_pred)
accuracy_score(y_train,y_pred)
```

```python
X_test = df_test[features]
display(X_test.head())
```

```python
y_pred = clf.predict(X_test)
print(y_pred)
print(len(y_pred))
```

```python
df_result[1]=y_pred
display(df_result.head())
```

```python
df_result.to_csv("./result.csv",index=False,header=None)
```
