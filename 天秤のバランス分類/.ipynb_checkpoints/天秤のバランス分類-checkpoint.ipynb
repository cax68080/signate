```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as LR 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import math
```

```python
df_train = pd.read_csv("./train.tsv",sep="\t")
df_test = pd.read_csv("./test.tsv",sep="\t")
sample = pd.read_csv("sample_submit.csv", header=None)
display(df_train.head())
#print(df_train.shape)
#display(df_train.dtypes)
display(df_test.head())
#print(df_test.shape)
#display(df_test.dtypes)
display(sample.head())
```

```python
df_train["left_point"] = df_train["left_weight"] * df_train["left_distance"]
df_train["right_point"] = df_train["right_weight"] * df_train["right_distance"]
df_test["left_point"] = df_test["left_weight"] * df_test["left_distance"]
df_test["right_point"] = df_test["right_weight"] * df_test["right_distance"]
display(df_train.head())
display(df_test.head())
```

```python
target_col = ["class"]
feature_col = ["left_weight","left_distance","right_weight","right_distance"]
df_y = df_train[target_col]
df_X = df_train[feature_col]
X_train,X_test,y_train,y_test = train_test_split(df_X,df_y,random_state=0,test_size=0.2)
print(X_train.shape)
print(y_train.shape)
model = LR()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
y_class = []
for i in y_pred:
    #print(i)
    for j in i:
        y_class.append(math.floor(abs(j)))
print(y_class)
#print(y_pred)
```

```python
accuracy = accuracy_score(y_class,y_test)
print(accuracy)
```
