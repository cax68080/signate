```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df_train = pd.read_csv("./train.csv")
df_test = pd.read_csv("./test.csv")
df_result = pd.read_csv("./sample_submission.csv",header=None)
display(df_train.head())
display(df_test.head())
display(df_result.head())
```

```python
train_dummies=pd.get_dummies(df_train[["sex","smoking","diabetes","anaemia","high_blood_pressure"]])
display(train_dummies.head())
```

```python
target_col =["target"]
feature_col = ["sex","smoking","diabetes","high_blood_pressure","creatinine_phosphokinase","platelets","serum_creatinine"]
y_train = df_train[target_col]
X_train = df_train[feature_col]
display(y_train)
display(X_train)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
df_train["pred_target"] = model.predict(X_train)
display(df_train[["target","pred_target"]].head())
```

```python
accuracy = accuracy_score(df_train["target"],df_train["pred_target"])
print(accuracy)
```

```python
X_test = df_test[feature_col]
display(X_test)
```

```python
df_result[1] = model.predict(X_test)
display(df_result.tail())
```

```python
df_result.to_csv("./result.csv",header=None,index=False)
```
