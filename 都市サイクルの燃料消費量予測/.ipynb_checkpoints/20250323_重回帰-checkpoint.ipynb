```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df_train = pd.read_csv("train.tsv",sep="\t")
df_test = pd.read_csv("test.tsv",sep="\t")
df_result = pd.read_csv("sample_submit.csv",header=None)
df_train = df_train.drop(df_train.columns[[0]],axis=1)
df_test = df_test.drop(df_test.columns[[0]],axis=1)
display(df_train.head())
display(df_test.head())
display(df_result.head())
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
df_train[df_train["horsepower"].apply(lambda s:pd.to_numeric(s, errors='coerce')).isnull()]
```

```python
df_test[df_test["horsepower"].apply(lambda s:pd.to_numeric(s, errors='coerce')).isnull()]
```

```python
df_train_wk = df_train.drop([25,80,176])
df_train_wk.shape
df_train_wk["horsepower"] = df_train_wk["horsepower"].astype("float64")
df_train_wk.dtypes
```

```python
hp_mean = df_train_wk["horsepower"].mean()
print(hp_mean)
```

```python
df_train["horsepower"] = pd.to_numeric(df_train["horsepower"],errors='coerce')
print(df_train['horsepower'].isnull().sum())
```

```python
df_train["horsepower"] = df_train["horsepower"].fillna(hp_mean)
df_train["horsepower"]=df_train["horsepower"].astype("float64")
print(df_train.info())
print(df_train['horsepower'].isnull().sum())
```

```python
df_test_wk = df_test.drop([50,85,181])
df_test_wk.shape
df_test_wk["horsepower"] = df_test_wk["horsepower"].astype("float64")
df_test_wk.dtypes
```

```python
hp_mean = df_test_wk["horsepower"].mean()
print(hp_mean)
```

```python
df_test["horsepower"] = pd.to_numeric(df_test["horsepower"],errors='coerce')
print(df_test['horsepower'].isnull().sum())
```

```python
df_test["horsepower"] = df_test["horsepower"].fillna(hp_mean)
df_test["horsepower"]=df_test["horsepower"].astype("float64")
print(df_test.info())
print(df_test['horsepower'].isnull().sum())
```

```python
X_train = df_train.drop(["mpg"],axis=1)
y_train = df_train["mpg"]
display(X_train.head())
display(y_train.head())
```

```python
display(X_train.head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_cols = X_train.columns[1:4]
scaler.fit(X_train[num_cols])
scaler.fit(df_test[num_cols])

X_train[num_cols] = scaler.transform(X_train[num_cols])
df_test[num_cols] = scaler.transform(df_test[num_cols])
display(X_train.iloc[:2])
display(df_test.iloc[:2])
```

```python
X_train = X_train.drop(["car name"],axis=1)
X_test = df_test.drop(["car name"],axis=1)
display(X_train.head())
display(X_test.head())
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,y_train)
print(model.get_params())
```

```python
y_train_pred = model.predict(X_train)
print("RMSE test: %.2f" % (mean_squared_error(y_train,y_train_pred) ** 0.5))
```

```python
y_test_pred = model.predict(X_test)
y_test_pred
```

```python
df_result[1] = y_test_pred
display(df_result.head())
```

```python
df_result.to_csv("result.csv",index=False,header=None)
```
